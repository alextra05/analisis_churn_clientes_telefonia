AQUI LO QUE VOY A HACER ES HACER UNA MARCA DE CALOR PARA SABER QUE VARIABLES ESTAN RELACCIONADAS ENTRE SI DE TIPO NUMERICO(INT64 Y FLOAT64)


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Leer el archivo CSV
df = pd.read_csv('billing_clients.csv')

# Seleccionar columnas con tipos de datos int64 y float64A
columnas_interes = df.select_dtypes(include=['int64', 'float64']).columns

# Calcular la matriz de correlaciones solo para las columnas de interés
corr_matrix = df[columnas_interes].corr()

# Visualizar la matriz de correlaciones con un mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de Correlaciones')
plt.show()
```

AQUÍ LO QUE VOY A HACER ES PROPONER DIFERENTES PROPUESTAS DE RETENCIÓN PARA LOS CLIENTES:

Promoción de Método de Pagos Automáticos:

Voy a ofrecer descuentos de entre un 5-10% en los primeros 6 meses para quienes migren a métodos de pago automáticos.

Creación de Paquetes Combinados de Soporte Técnico y Seguridad:

Crear ofertas para paquetes con descuento que incluyan los siguientes servicios: Online Security, Online Backup y Tech Support.

Ofertas de Retención:

Enviar una notificación a clientes que alcanzaron los 2 meses de permanencia (punto promedio del churn) con promociones para extender contratos.

```python
import pandas as pd

# Leer el archivo CSV
df = pd.read_csv('billing_clients.csv')

# Filtrar los clientes que se han ido (aquellos con churn_date no nulo)
churners = df[df['churn_date'].notnull()]

# Identificar clientes que se han ido con métodos de pago manuales
manual_payment_churners = churners[churners['paymentmethod'].isin(['Mailed check', 'Electronic check'])]

# Calcular el 25% de estos clientes
retained_clients_count = int(manual_payment_churners.shape[0] * 0.25)

# Aplicar una promoción: descuento del 5-10% durante los primeros 6 meses si cambian a métodos automáticos
manual_payment_churners = manual_payment_churners.copy()
manual_payment_churners.loc[:, 'promotional_offer'] = manual_payment_churners['monthlycharges'] * 0.05

# Seleccionar el primer 25% de clientes para la promoción
promotional_clients = manual_payment_churners.head(retained_clients_count)

# Calcular el ahorro total para los clientes
total_savings = promotional_clients['promotional_offer'].sum() * 6

# Calcular los ingresos totales para la empresa (asumiendo que los clientes pagan el 95% restante de sus cargos)
total_revenue = (promotional_clients['monthlycharges'].sum() * 0.95) * 6

# Mostrar los resultados
print(f"{manual_payment_churners.shape[0]} clientes que se han ido tenían métodos de pago manuales.")
print(f"De estos, captamos al 25%, que son {retained_clients_count} clientes.")
print(f"A estos clientes se les ofreció un descuento del 5-10% durante los primeros 6 meses si cambian a métodos de pago automáticos.")
print(f"El ahorro total para estos clientes es: ${total_savings:.2f}")
print(f"Los ingresos totales para la empresa de estos clientes retenidos son: ${total_revenue:.2f}")
```

```python
import pandas as pd

# Leer el archivo CSV
df = pd.read_csv('billing_clients.csv')

# Filtrar los clientes que se han ido (aquellos con churn_date no nulo)
churned_clients = df[df['churn_date'].notnull()]

# Identificar clientes que se fueron cerca del punto crítico (2.25 meses ± 0.5 meses)
critical_churners = churned_clients[
    (churned_clients['tenure_months'] >= 1.75) &  # 2.25 - 0.5
    (churned_clients['tenure_months'] <= 2.75)    # 2.25 + 0.5
    ]

# Lista de servicios según las columnas de tu CSV
services = [
    'internetservice',
    'onlinesecurity',
    'onlinebackup',
    'deviceprotection',
    'techsupport',
    'phone_lines',
    'streaming'
]

# Analizar los servicios que tenían
churner_analysis = critical_churners[['customerid', 'tenure_months', 'monthlycharges'] + services].copy()

# Identificar qué servicios les faltaban antes de irse
churner_analysis['missing_services'] = churner_analysis[services].apply(
    lambda row: [service for service in services if row[service] == 'No'],
    axis=1
)

# Calcular métricas
total_churners = len(churned_clients)
critical_period_churners = len(critical_churners)
lost_monthly_revenue = critical_churners['monthlycharges'].sum()
lost_annual_revenue = lost_monthly_revenue * 12

# Mostrar resumen
print("\nANÁLISIS DE CLIENTES QUE CANCELARON")
print("---------------------------------")
print(f"Total de clientes que cancelaron: {total_churners}")
print(f"Clientes que cancelaron cerca de 2.25 meses: {critical_period_churners}")
print(f"Porcentaje del total: {(critical_period_churners/total_churners)*100:.2f}%")
print(f"Ingresos mensuales perdidos: ${lost_monthly_revenue:,.2f}")
print(f"Ingresos anuales perdidos: ${lost_annual_revenue:,.2f}")

# Mostrar análisis detallado
print("\nDETALLE DE CLIENTES QUE CANCELARON EN PERIODO CRÍTICO:")
churner_analysis[['customerid', 'tenure_months', 'monthlycharges', 'missing_services']]

```

```python
# Análisis detallado de cada servicio faltante
print("\nDETALLE DE SERVICIOS FALTANTES:")
for service in services:
    missing_count = churner_analysis[churner_analysis[service] == 'No'].shape[0]
    percentage = (missing_count / critical_period_churners) * 100
    print(f"{service}: {missing_count} clientes ({percentage:.1f}%)")
# Analizar patrones de servicios faltantes
print("\nDISTRIBUCIÓN DE SERVICIOS FALTANTES AL MOMENTO DE CANCELAR:")
service_gaps = churner_analysis['missing_services'].apply(len).value_counts().sort_index()
service_gaps
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leo el archivo CSV
data = pd.read_csv('billing_clients.csv')

# Me aseguro de trabajar directamente con un DataFrame limpio
data_filtered = data.copy()  # Evito problemas de vistas o copias

# Convierto las columnas churn_date y birth_date a tipo datetime con manejo de errores
data_filtered['churn_date'] = pd.to_datetime(data_filtered['churn_date'], errors='coerce')
data_filtered['birth_date'] = pd.to_datetime(data_filtered['birth_date'], errors='coerce')

# Elimino las filas con fechas inválidas
data_filtered = data_filtered.dropna(subset=['churn_date', 'birth_date'])

# Calculo la columna edad_de_abandono
data_filtered['edad_de_abandono'] = (data_filtered['churn_date'] - data_filtered['birth_date']).dt.days // 365

# Gráfico 1: Clientes que se han ido por género y estado civil
churners = data_filtered[data_filtered['churn_date'].notnull()]
gender_marital_counts = churners.groupby(['gender', 'married']).size()
labels1 = [f'{gender} {"Casado" if married == "Yes" else "No casado"}' for (gender, married) in gender_marital_counts.index]

# Gráfico 2: Edad de abandono de los clientes
bins = range(0, 100, 5)

# Configuro el tamaño de la figura y el layout
plt.figure(figsize=(18, 6))

# Subplot 1
plt.subplot(1, 2, 1)
plt.pie(gender_marital_counts, labels=labels1, autopct='%1.1f%%', startangle=140)
plt.title('Clientes que se han ido por género y estado civil')

# Subplot 2
plt.subplot(1, 2, 2)
plt.hist(data_filtered['edad_de_abandono'], bins=bins, color='skyblue', edgecolor='black', linewidth=1.2)
plt.title('Edad de abandono de los clientes')
plt.xlabel('Edad')
plt.ylabel('Número de clientes')
plt.xticks(bins)

# Ajusto el layout y muestro los gráficos
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leo el archivo CSV
data = pd.read_csv('billing_clients.csv')

# Gráfico 1: Clientes solteros que se dieron de baja con fibra óptica y pago electrónico
churners = data[data['churn_date'].notnull()]
churners_single = churners[churners['married'] == 'No']
churners_single_fiber = churners_single[churners_single['internetservice'] == 'Fiber optic']
churners_single_fiber_electronic = churners_single_fiber[churners_single_fiber['paymentmethod'] == 'Electronic check']

churners_single_fiber_electronic_count = churners_single_fiber_electronic.shape[0]
print(f'Clientes que se han dado de baja, son solteros, tienen contratado fibra óptica y utilizan el método de pago Electronic check: {churners_single_fiber_electronic_count} clientes')

services = ['onlinesecurity', 'onlinebackup', 'deviceprotection', 'techsupport']
service_counts = {service: churners_single_fiber_electronic[service].value_counts() for service in services}

for service in services:
    for status, count in service_counts[service].items():
        status_text = "" if status == "Yes" else "no "
        print(f'Clientes que se han dado de baja, son solteros, tienen contratado fibra óptica, utilizan Electronic check, y {status_text}tienen contratado {service}: {count} clientes')

# Gráfico 2: Clientes que se han ido por servicio de internet
internet_service_counts = churners['internetservice'].value_counts()
labels2 = internet_service_counts.index

# Configuro el tamaño de la figura y el layout
fig, axs = plt.subplots(1, 2, figsize=(18, 6))

# Subplot 1: Clientes solteros con fibra óptica y pago electrónico
colors = {'Yes': 'green', 'No': 'red'}
bar_width = 0.4
x_positions = range(len(services))

for i, service in enumerate(services):
    counts = service_counts[service]
    yes_count = counts.get('Yes', 0)
    no_count = counts.get('No', 0)
    axs[0].bar(i - bar_width / 2, no_count, width=bar_width, color=colors['No'], label='No' if i == 0 else "")
    axs[0].bar(i + bar_width / 2, yes_count, width=bar_width, color=colors['Yes'], label='Yes' if i == 0 else "")

axs[0].set_xticks(x_positions)
axs[0].set_xticklabels(services)
axs[0].set_ylabel('Número de Clientes')
axs[0].set_title('Clientes Solteros que se Dieron de Baja con Fibra Óptica y Pago Electrónico por Estado del Servicio')
axs[0].legend(title='Estado del Servicio')

# Subplot 2: Clientes que se han ido por servicio de internet
axs[1].pie(internet_service_counts, labels=labels2, autopct='%1.1f%%', startangle=140)
axs[1].set_title('Clientes que se han ido por servicio de internet')

# Ajusto el layout y muestro los gráficos
plt.tight_layout()
plt.show()
```
